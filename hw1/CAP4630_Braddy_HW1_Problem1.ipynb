```
from mpl_toolkits import mplot3d
from mpl_toolkits.mplot3d import axes3d
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
from matplotlib import cm

fig = plt.figure()
ax = plt.axes(projection='3d')
```

Setting Up Function :: Must uncomment the function you wish to use then run the code under either Function 1, 2, or 3 respectively

```
def f(x, y):
    #return np.sin(0.09 * x) + np.cos(0.05 * y)
    #return np.cos(0.09 * x) + np.sin(0.05 * y)
    return np.sin(0.09 * y) + np.cos(0.05 * x) * 2
```

```
x = np.linspace(-40, 40, 40)
y = np.linspace(-40, 40, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

plt.contour(X, Y, Z, colors='black');
```

```
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar();
```

Function 1: sin(0.09 * x) + cos(0.05 * y)


```
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');


xmin, ymin = np.unravel_index(np.argmin(Z), Z.shape)
xmax, ymax = np.unravel_index(np.argmax(Z), Z.shape)

mi = (X[xmin,ymin], Y[xmin,ymin], Z.min())
ma = (X[xmax, ymax], Y[xmax, ymax], Z.max())

Ami = np.array([mi] * 4)
Ama = np.array([ma] * 4)
for i, v in enumerate([-40,40,-2]):
     Ami[i,i] = v 
     Ama[i,i] = v 

AmiTwo = np.array([mi] * 2)
AmaTwo = np.array([ma] * 2)
Grad = np.array([mi] * 1)

AmiTwo[0] = Ami[1]
AmaTwo[0] = Ama[0]
AmiTwo[1] = Ami[3]
AmaTwo[1] = Ama[3]
Grad[0] = -17.43589744 ,-1.02564103, .09557635

#plot points.
print("Local Minima: ")
print(Ami[1])
print(Ami[3])
print("Local Maxima: ")
print(Ama[1])
print(Ama[3])
print("Gradient = 0 : ")
print(Grad[0])

ax.plot(AmiTwo[:,0], AmiTwo[:,1], AmiTwo[:,2], marker="o", ls="", c=cm.coolwarm(0.))
ax.plot(AmaTwo[:,0], AmaTwo[:,1], AmaTwo[:,2], marker="o", ls="", c=cm.coolwarm(1.))
ax.plot(Grad[:,0], Grad[:,1], Grad[:,2], marker="o", ls="", c=cm.coolwarm(0.))

ax.view_init(60, 85)
```
